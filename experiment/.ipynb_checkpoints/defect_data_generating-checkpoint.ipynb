```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold

import warnings
warnings.filterwarnings('ignore')
```

```python
def dust_cover(data):
    
    d_data = []
    for i in data:
        d_data.append(i*(np.random.uniform(0.5,0.9)))
        
    mse = mean_squared_error(data, d_data)
    return d_data

def normal_noise_data(data):
    
    d_data = []
    for i in data:
        d_data.append(i*(1+np.random.normal(0,0.1)))
    return d_data

def open_circuit(data):
    
    d_data = []
    break_p = np.random.randint(0.1*len(data),0.9*len(data))
    
    r = [0,2,3,4,5]
    level = random.choice(r)
    for i, i_data in enumerate(data):
        if i < break_p:
            d_data.append(i_data*(1+np.random.normal(0,0.02)))
        else:
            if level == 0:
                d_data.append(np.random.normal(0,0.02))
            else:
                d_data.append(i_data*(1/level)*(1+np.random.normal(0,0.02)))
     
    mse = mean_squared_error(data, d_data)
    if mse < 5:
        return open_circuit(data)
    else:
        return d_data

def partial_shading(data):
    
    d_data = []
    
    num_partial_p = np.random.randint(2,8)

    percent = []
    for i in range(num_partial_p):
        if i % 2 == 0:
            level = np.random.uniform(0.0,0.5)
        else:
            level = 1
        percent.append(level)
        
    points = []
    for i in range(num_partial_p):
        points.append(np.random.randint(0,len(data)))
    points.sort()

    partial = {point:level for point, level in zip(points,percent)}
    
    d_data = []
    level = 1
        
    for j,j_data in enumerate(data):
        
        if j in partial:
            level = partial[j]            
        d_data.append(j_data*level*(1+np.random.normal(0,0.05))) 
      
    mse = mean_squared_error(data, d_data)
    if mse < 5:
        return partial_shading(data)
    else:
        return d_data
    
def create_dataset(data):
        
#     data = data.values
    dataX, dataY = [],[]
    for i in range(len(data)):
        dataX.append(data[i][:-1])
        dataY.append(int(data[i][-1])) 
            
    return np.array(dataX), np.array(dataY)

def random_forest_class(dataset):
    dataset = dataset.values
        
    X, Y = create_dataset(dataset)
    cvscores = []
    cm = []
    kfold = StratifiedKFold(n_splits=3, shuffle=True)
        
    for train_index, test_index in kfold.split(X, Y):
            
        X_train = X[train_index]
        y_train = Y[train_index]
        X_test = X[test_index]
        y_test = Y[test_index]
        
        rf_model = RandomForestClassifier(max_depth=10, random_state=1).fit(X_train, y_train)
        rf_prediction = rf_model.predict(X_test)
        cvscores.append(accuracy_score(y_test, rf_prediction)*100)
        cm.append(confusion_matrix(y_test, rf_prediction))
        
    return [cvscores, cm, rf_model,X_test]
        
def visualize(data,d_data=None,title_name=None,save=False):
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1)

    fig_1 = plt.plot(data)
    fig_2 = plt.plot(d_data) 
    
    ax.set_title("Simulated Output Under " + title_name + " Condition")
    ax.set_xticks(np.linspace(0,len(data),12))
    ax.xaxis.set_ticklabels(['6am','7am','8am','9am','10am','11am','12pm','1pm','2pm','3pm','4pm','5pm','6pm'])
    ax.set_xlabel('Time',fontsize=14)
    ax.set_ylabel('Power(W)',fontsize=14)
    plt.legend(["Normal Output", "Anomaly Output"])
    
    plt.tight_layout()
    if save is True:
        plt.savefig(title_name+'.jpg', format='jpg', dpi=1000)
    plt.show()
    
def visualize_2(data,d_data=None,title_name=None,save=False):
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1)

    fig_1 = plt.plot(data)
    fig_2 = plt.plot(d_data) 
    
    ax.set_title("Open Circuit Observation for Parallel Connected Setup")
    ax.set_xticks(np.linspace(0,len(data),6))
    ax.xaxis.set_ticklabels(['0 min','2 mins','4 mins','6 mins','8 mins','10 mins'])
    ax.set_ylim([0, max(data)*1.2])
    ax.set_xlabel('Time',fontsize=14)
    ax.set_ylabel('Power(W)',fontsize=14)
    plt.legend(["Normal Output", "Open Circuit"])
    
    plt.tight_layout()
    if save is True:
        plt.savefig(title_name+'.jpg', format='jpg', dpi=1000)
    plt.show()
    
    
def generate_anomaly(dataset, save=False):
    
    dataset = dataset.values
    feature = []
    s_size = 50
    
    # Daily loop
    for i in range(len(dataset[0])):
    
        # Load per day data
        data = dataset[:,i]
    
        # Drop NaN
        data = [j for j in data if not np.isnan(j)]
        # Generate Index
        index = [i for i in range(len(data))]
        
        # Skip the data if the daily output is None or too small
        if len(data) == 0 or (max(data)-min(data)) < 10:
            continue
    
        # Skip the data if its size is less than sample size
        if len(data) < s_size:
            continue
        else:
            # Uniformly sample the index until the number of data reach to sample size
            while True:
                samples = random.sample(index,s_size)
                if len(samples) == s_size:
                    samples.sort()
                    break
        # Use sampled index to create sampled data
        data = [data[k] for k in samples]

        anomaly_list = ["partial","cover","open","normal_noise"]
        
        for anomaly in anomaly_list:
    
            # Simulate anomaly data
            if anomaly is "partial":
                d_data = partial_shading(data)
                
            elif anomaly is "cover":
                d_data = dust_cover(data)

            elif anomaly is "open":
                d_data = open_circuit(data)

            else:
                d_data = normal_noise_data(data)

    
            # Compute feature and save the fature data in csv file
            diff = []
            for g in range(s_size):
                if data[g] == 0:
                    data[g] = 1
                diff.append((data[g]-d_data[g])/data[g])
                
            diff.append(anomaly_list.index(anomaly))
            feature.append(diff)
            
    # Convert list into DataFrame
    feature_data = pd.DataFrame(feature) 
    
    # Save the feature data as csv file
    if save is True:
        export_csv = feature_data.to_csv ("features.csv", index = True, header=True)
        
    return feature_data

def confusion_matrix_plot(cm,acc,save=False):
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1) 
    sns.set(font_scale=1)
    sns.heatmap(cm, annot=True,fmt='g',annot_kws={"size": 15},cmap="Blues", square=True, cbar=False)

    # labels, title and ticks
    ax.set_xlabel('Prediction',fontsize=18)
    ax.set_ylabel('Ground Truth',fontsize=18)
    title_name = 'Confusion Matrix - Accuracy ' + str(round(acc,2)) + "%"
    ax.set_title(title_name,fontsize=20) 
    ax.xaxis.set_ticklabels(['Partial Shading', 'Object Covering', 'Wiring Anomaly','Normal Condition'],rotation=45)
    ax.yaxis.set_ticklabels(['Partial Shading', 'Object Covering', 'Wiring Anomaly','Normal Condition'],rotation=45)

    plt.tight_layout()
    if save is True:
        plt.savefig(title_name+'.jpg', format='jpg', dpi=1000)
    plt.show()
```

```python
## Create anomaly dataset
dataset = pd.read_csv('28563614.csv', index_col=0)
generate_anomaly(dataset,True)
```

```python
# import helper function
from sklearn.utils import resample
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor

# prepare the data for regression
def create_dataset(data):
    dataX, dataY = [],[]
    for i in range(len(data)):
        dataX.append(data[i][:-1])
        dataY.append([data[i][-1]])
        
    return np.array(dataX), np.array(dataY)


# Mean Absolute Percentage Error
def mean_absolute_percentage_error(data_true, data_predict):
    error = 0
    count = 0
    data_true_de = sum(data_true)/len(data_true)
    
    for i in range(len(data_true)):
        error += np.abs((data_true[i]-data_predict[i])/data_true_de)

    return((error/len(data_true))*100)


def random_forest_reg(X,Y):
    
    cvscores = []
    k_fold = KFold(n_splits=4, shuffle=True, random_state=0)
        
    for train_index, test_index in k_fold.split(X, Y):
        
        X_train = X[train_index]
        y_train = Y[train_index]
        
        X_test = X[test_index]
        y_test = Y[test_index]        
        
        rf_model = RandomForestRegressor(n_estimators=10,max_depth=10, random_state=0).fit(X_train, y_train)
        rf_prediction = rf_model.predict(X_test)
        score = mean_absolute_percentage_error(y_test, rf_prediction)
        cvscores.append(score)
        
    return np.mean(cvscores)

def regression(dataset):
    dataset = dataset.values
    
    # normalize the data
    scaler = MinMaxScaler(feature_range=(0,1))
    dataset = scaler.fit_transform(dataset)
    
    X,Y = create_dataset(dataset)
    
    return random_forest_reg(X,Y)
```

```python
dataset = pd.read_csv('case_study_week.csv', index_col=0)

print(regression(dataset))

day = 912
day = random.choice([index for index in set(dataset.index)])

data = dataset.loc[day]['28563614'].values
d_data = open_circuit(data)

replace_data = []
for index in set(dataset.index):

    if index == day:
        replace_data.extend(d_data)
    else:
        replace_data.extend(dataset['28563614'].loc[index].values)
        
dataset['anomaly'] = replace_data

print(regression(dataset))
```
