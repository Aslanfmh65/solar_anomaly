```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import helper
```

```python
def dust_cover(data):
    
    d_data = []
    levels = []
    
    leaf_power_dict = {1:0.85,2:0.7,3:0.55,4:0.4,5:0.25}
    
    leaf_num = np.random.randint(1,6)
    
    level = leaf_power_dict[leaf_num]
    
    for i in data:
        noise = np.random.normal(-0.05,0.05)
        noise_level = level*(1+noise)        
        levels.append(noise_level)
        d_data.append(i*noise_level)
    return d_data, levels

def normal_noise_data(data):
    
    d_data = []
    for i in data:
        d_data.append(i*(1++np.random.normal(0,0.01)))
    return d_data

def open_circuit(data):
    
    d_data = []
    break_p = np.random.randint(0.2*len(data),0.7*len(data))
    parallel = random.choice([0,1/2,1/3])
    
    levels = []
    
    for i, i_data in enumerate(data):
        if i < break_p:
            levels.append(1)
            d_data.append(i_data*(1+np.random.normal(0,0.01)))
        else:
            d_data.append(i_data*parallel*(1+np.random.normal(0,0.01)))
            levels.append(parallel*(1+np.random.normal(0,0.01)))
            
    return d_data, levels

def partial_shading(data):
    
    d_data = []
    
    num_partial_p = np.random.randint(2,6)

    percent = []
    for i in range(num_partial_p):
        if i % 2 == 0:
            level = np.random.uniform(0.1,0.5)
        else:
            level = 1
        percent.append(level)
        
    points = []
    for i in range(num_partial_p):
        points.append(np.random.randint(0,len(data)))
    points.sort()

    partial = {point:level for point, level in zip(points,percent)}
    
    d_data = []
    factor = []
    level = 1
        
    for j,j_data in enumerate(data):
        
        if j in partial:
            level = partial[j]            
        d_data.append(j_data*level*(1+np.random.normal(0,0.01)))  
        factor.append(level)
    
    return d_data,factor

def visualize(data, dpi, save=False):
                    
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(1,1,1)
    
    plt.plot(data)
    plt.legend(['Normal Data','Synthetic Data'])

    title = 'Synthetic Power Output Under Partial Shaded '
    
    ax.set_title(title,fontsize=25,y=1.05)
    
    # Time x-axis(bottom)
    ax.set_xticks(np.linspace(0,len(data),7))
    ax.xaxis.set_ticklabels(['7:00','9:00','11:00','13:00','15:00','17:00','19:00'], fontsize=12)
    ax.set_xlabel('Time',fontsize=20)
    
    ax.set_ylabel('Power (W)',fontsize=20)
    
    if save == True:
    
        plt.savefig(title + '.jpg', format='jpg', dpi=dpi, bbox_inches='tight')
```

```python
dataset = pd.read_csv('east_data_clean.csv')
defect_data = dataset[dataset['correlation']=='low']
dataset = dataset[dataset['label']=='normal']

day_list = list(set(dataset['date']))

drop_col = [d_col for d_col in dataset.columns if d_col[:5] != 'panel' and d_col != 'date']
dataset = dataset.drop(drop_col,axis=1)
defect_data = defect_data.drop(drop_col,axis=1)
panels = [i for i in dataset.columns if i != 'date']
```

```python
# Create Defect dataset

store_dict = {}

while day_list:
    
    df = pd.DataFrame(columns=panels)
    defect_col = []
    date_col = []
    
    random.shuffle(day_list)
    day = day_list.pop()
    d_panel = random.choice(panels)
    
    for p in panels:
        if p != d_panel:
            df[p] =  dataset[dataset['date']==day][p]
    
    data = dataset[dataset['date']==day].drop(['date'],axis=1)[d_panel].values
    cond = random.choice(['open','cover','normal'])
    
    if cond == 'open':
        d_data,_ = open_circuit(data)
        defect_col = ['open']*len(d_data)
        
    elif cond == 'cover':
        d_data,_ = dust_cover(data)
        defect_col = ['cover']*len(d_data)
        
    else:
        d_data = normal_noise_data(data)
        defect_col = ['normal']*len(d_data)
        
    df[d_panel] = d_data
    df['defect'] = defect_col
    df['date'] = [day]*len(d_data)
    
    if day not in store_dict:
        store_dict[day] = df
        
df = pd.DataFrame(columns=panels)
for i in store_dict:
    df = df.append(store_dict[i])
    
defect_data['defect'] = ["production_issue"]*len(defect_data)
df = df.append(defect_data)
```

```python
day_list = list(set(df['date']))
```

```python
res = pd.DataFrame()
labels = []
stds = []
dates = []

for day in day_list:

    data = df[df['date']==day]
    label = data['defect'].values[0]
    data = data.drop(['date','defect'],axis=1)

    cols = list(data.columns)
    diffs = []

    for i in range(len(cols)):
        col = cols[i]
        rest_cols = cols[:i]+cols[i+1:]
    
        test_data = data[col]
        mean_data = data[rest_cols].mean(axis=1)
    
        diff = sum(abs(test_data-mean_data))
    
        diffs.append(diff)
    
    if sum(diffs) < 1:
        stds.append(100)
        labels.append(2)
        dates.append(day)
        continue
    diffs = [i/sum(diffs) for i in diffs]
    
    stds.append(np.std(diffs))
    dates.append(day)
    if label == 'normal':
        labels.append(1)
    elif label == 'production_issue':
        labels.append(2)
    else:
        labels.append(0)

res['date'] = dates
res['std'] = stds
res['label'] = labels
```

```python
import helper
model = helper.modeling()
scores, cm, rf_model,X_test,y_test,test_date = model.random_forest_class(res['label']!=2)
scores
```

```python
import os
import pywt
import keras
import itertools
import numpy as np
import pandas as pd
import seaborn as sns
import numpy.fft as fft
from scipy import signal
from sklearn.svm import SVC 
from scipy.stats import skew
from scipy.fftpack import dct
from keras.layers import LSTM
from keras.layers import Dense
from scipy.stats import entropy
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
from keras.utils import np_utils
from keras.layers import Flatten
from keras.layers import Dropout
from keras.optimizers import SGD
from keras.models import Sequential
from keras.layers import Activation
from scipy.signal import find_peaks
from keras.utils import to_categorical
from mpl_toolkits.mplot3d import Axes3D
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from keras.layers.convolutional import Conv1D
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from keras.layers import GlobalAveragePooling1D
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from keras.layers.convolutional import MaxPooling1D
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import StratifiedKFold


import warnings
warnings.filterwarnings('ignore')
```

```python
def create_dataset(data):
        
    data = data.values
    dataX, dataY = [],[]
    for i in range(len(data)):
        dataX.append(data[i][:-1])
        dataY.append(int(data[i][-1])) 
            
    return np.array(dataX), np.array(dataY)

def random_forest(dataset):
        
    X, Y = create_dataset(dataset)
    cvscores = []
    cm = []
    kfold = StratifiedKFold(n_splits=3, shuffle=True)
        
    for train_index, test_index in kfold.split(X, Y):
            
        X_train = X[train_index]
        y_train = Y[train_index]
        X_test = X[test_index]
        y_test = Y[test_index]
        
        rf_model = RandomForestClassifier(max_depth=10, random_state=1).fit(X_train, y_train)
        rf_prediction = rf_model.predict(X_test)
        cvscores.append(accuracy_score(y_test, rf_prediction)*100)
        cm.append(confusion_matrix(y_test, rf_prediction))
        
    return [cvscores, cm]

def confusion_matrix_plot(cm,acc,save=False):
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1) 
    sns.set(font_scale=1)
    sns.heatmap(cm, annot=True,fmt='g',annot_kws={"size": 25},cmap="Blues", square=True, cbar=False)

    # labels, title and ticks
    ax.set_xlabel('Prediction',fontsize=30)
    ax.set_ylabel('Ground Truth',fontsize=30)
    title_name = 'Condition Classification Acc ' + str(round(acc,2)) + "%"
    ax.set_title(title_name,fontsize=30,y=1.2) 
#     ax.xaxis.set_ticklabels(['Strong', 'Moderate', 'Low'],fontsize=18,rotation=45)
#     ax.yaxis.set_ticklabels(['Strong', 'Moderate', 'Low'],fontsize=18,rotation=45)
    ax.xaxis.set_ticklabels(['Partial Shading', 'Object Covering', 'Open Circuit','Normal Condition'],fontsize=16,rotation=45)
    ax.yaxis.set_ticklabels(['Partial Shading', 'Object Covering', 'Open Circuit','Normal Condition'],fontsize=16,rotation=45)

    plt.tight_layout()
    if save is True:
        plt.savefig(title_name+'.jpg', format='jpg', dpi=1000)
    plt.show()
```

```python
dataset = pd.read_csv('defect_data.csv', index_col=0)
score,cm = random_forest(dataset)
score
```
