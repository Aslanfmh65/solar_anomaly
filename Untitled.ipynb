```python
def create_dataset(data):
    dataX, dataY = [],[]
    for i in range(len(data)):
        dataX.append(data[i][:-1])
        dataY.append([data[i][-1]])
        
    return np.array(dataX), np.array(dataY)

def generate_mape_plot(dataset,save=False):

    results = []
    for corr in ['strong','moderate','low']: 
        data = dataset[dataset['correlation']==corr]
        drop_col = [d_col for d_col in data.columns if d_col[:5] != 'panel']
        data = data.drop(drop_col,axis=1)
        res, _ = model.model_predict(data)
        results.append(res)
        
        
    # make the plot
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1)
    
    title_name = "Power Prediction Error - East Roof Dataset"
    box = plt.boxplot(results,patch_artist=False, showmeans=True)
    
    ax.set_title(title_name,fontsize=20)
    ax.xaxis.set_ticklabels(["Strong", "Moderate", "Low"], fontsize=15)
    ax.set_xlabel('Correlation',fontsize=20)

    power_y = [0,5,10,15,20,25,30,35,40]
    ax.set_yticks(power_y)
    ax.yaxis.set_ticklabels(power_y, fontsize=15)
    ax.set_ylabel('MAPE(%)',fontsize=20)
    
    if save == True:
        plt.savefig(title_name + '.jpg', format='jpg', dpi=500, bbox_inches='tight')
        
def make_plot(results,save=False):
    # make the plot
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(1,1,1)
    
    title_name = "Power Prediction Error - East Roof Dataset"
    box = plt.boxplot(results,patch_artist=False, showmeans=True)
    
#     ax.set_title(title_name,fontsize=20)
    ax.xaxis.set_ticklabels(["1 Day", "2 Days", "3 Days", '4 Days', '5 Days'], fontsize=20)
#     ax.set_xlabel('Correlation',fontsize=20)


    power_y = [0,2,4,6,8,10,12,14]
    ax.set_yticks(power_y)
    ax.yaxis.set_ticklabels(power_y, fontsize=20)
    ax.set_ylabel('Power(W)',fontsize=25)
    
    if save == True:
        plt.savefig('size_experiment.jpg', format='jpg', dpi=500, bbox_inches='tight')
    
        
def cond_filter(dataset,target_condition):
    
    day_list = list(set(dataset['date']))

    target_day = []
    corr_dict = {'strong':0,'moderate':0}

    for day in day_list:

        data = dataset[dataset['date']==day]
    
        if len(data) <= 5:
            continue
    
        weather = collections.Counter(data['weather_icon']).most_common()[0][0]

        if weather not in target_condition:
            continue
        else:
            target_day.append(day)
        
    drop_col = [d_col for d_col in dataset.columns if d_col[:5] != 'panel' and d_col not in ['date','correlation']]
    new_dataset = dataset.drop(drop_col,axis=1)
          
    df = np.array([0]*9)       
    for day in target_day:
        data = new_dataset[dataset['date']==day]
        
        correlation = data['correlation'].values[0]
        if correlation == 'strong':
            corr_dict['strong'] += 1
        else:
            corr_dict['moderate'] += 1
    
        data = data.drop(['date','correlation'],axis=1).values
    
        df = np.vstack((df, data))
        
    return pd.DataFrame(df), corr_dict

def confusion_plot(cm):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(1,1,1) 
    sns.set(font_scale=1)
    sns.heatmap(cm, annot=True,fmt='g',annot_kws={"size": 20},cmap="Blues", square=True, cbar=False)

    # labels, title and ticks
    ax.set_xlabel('Prediction',fontsize=25)
    ax.set_ylabel('Ground Truth',fontsize=25)
    # title_name = 'Confusion Matrix - Accuracy ' + str(round(acc,2)) + "%"
    # ax.set_title(title_name,fontsize=20) 
    ax.xaxis.set_ticklabels(['Snow Cover', 'Normal','Overcast'],fontsize=20)
    ax.yaxis.set_ticklabels(['Snow Cover', 'Normal','Overcast'],fontsize=20)

    plt.tight_layout()

    plt.savefig('per_panel_snow_detection.jpg', format='jpg', dpi=1000)
    plt.show()
```

```python
day_list = [day for day in set(dataset['date'])]
feature_power = {}
drop_col = [col for col in dataset.columns if col[:5]!='panel']

max_power = []
s_depth = []

df = pd.DataFrame()
df_date = []

for day in day_list:
    data = dataset[dataset['date']==day]
    
    if len(data) < 50:
        continue
        
    
    cols = [col for col in data.columns if col[:5]=='panel' and col not in ['solar-tk']]
    solar_tk = data['solar-tk'].values
    snow = np.mean(data['snow_depth'].values)
    
    data = data.drop(drop_col,axis=1)
    

    for col in cols:
        capacity = sum(solar_tk-data[col].values)/sum(solar_tk)
        
        if capacity < 0.5:
            
            # Generate 40 feature 
            feature_index = []
            while len(feature_index) < 40:
                index = np.random.randint(0,len(data))
                if index not in feature_index:
                    feature_index.append(index)
            feature_index.sort()
            
            feature = ((solar_tk-data[col].values)/solar_tk)[feature_index]
            
#             feature = data[col].values[feature_index]/solar_tk[feature_index]
            feature_power['-'.join(day.split('/'))+'-'+panel[:7]] = feature
            
            df = df.append(pd.Series(feature), ignore_index=True)
            df_date.append(day)  
            max_power.append(data.values.max())
            s_depth.append(snow)
```

```python
# Raw data snow defect detection

import seaborn as sns
from sklearn.metrics import confusion_matrix 

def confusion_plot(cm):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(1,1,1) 
    sns.set(font_scale=1)
    sns.heatmap(cm, annot=True,fmt='g',annot_kws={"size": 20},cmap="Blues", square=True, cbar=False)

    # labels, title and ticks
    ax.set_xlabel('Prediction',fontsize=25)
    ax.set_ylabel('Ground Truth',fontsize=25)
    # title_name = 'Confusion Matrix - Accuracy ' + str(round(acc,2)) + "%"
    # ax.set_title(title_name,fontsize=20) 
    ax.xaxis.set_ticklabels(['Anomaly', 'Normal'],fontsize=20)
    ax.yaxis.set_ticklabels(['Anomaly', 'Normal'],fontsize=20)

    plt.tight_layout()

    plt.savefig('anomaly_detection.jpg', format='jpg', dpi=1000)
    plt.show()

dataset = pd.read_csv('east_data_clean.csv').dropna()
dataset = dataset[dataset['label']!='unknown_issue']

day_list = list(set(dataset['date']))

# Create object
visual = helper.visualizer()
model = helper.modeling()
process = helper.data_processing()

# Generate features
date = []
mape = []
std = []
power = []
coeff = []
corr = []

for day in day_list:
    data = dataset[dataset['date']==day]
    
    if len(data) < 4:
        continue
    
    cor = data['correlation'].values[0]
    
    if cor == 'strong' or cor == 'moderate':
        corr.append(1)
    else:
        corr.append(0)
    
    drop_col = [d_col for d_col in data.columns if d_col[:5] != 'panel']
    data = data.drop(drop_col,axis=1)

    res,_ = model.model_predict(data,'RF')
    
    # Correlation feature
    corr_res = []
    first = data.columns[0]
    for col in data.columns:
        if col != first:
            corr_res.append(pearsonr(data[col].values,data[first].values)[0])
    coeff.append(min(corr_res))
    
    power_level = min([(320-max(data[i]))/320 for i in data.columns])
    
    date.append('l'+'/'+day)
    power.append(power_level)
    mape.append(np.mean(res))
    std.append(np.std(res))
    

feature = pd.DataFrame()
feature['date'] = date
feature['mape'] = mape
feature['std'] = std
feature['power'] = power
feature['coeff'] = coeff
feature['corr'] = corr

# # export_csv = feature.to_csv ("lower_features.csv", index = False, header=True) 
# feature = pd.read_csv('sy.csv')
# feature = feature[['date','power','std','coeff','corr']]

df = pd.DataFrame()
df_2 = pd.DataFrame()

df_3 = pd.DataFrame()

# thresh = 0.8
# df = df.append(feature[feature['corr']==1][feature['power']<thresh])
# df = df.append(feature[feature['corr']==0][feature['power']>=thresh])

# df_2 = df_2.append(feature[feature['corr']==1][feature['power']>=thresh])
# df_2 = df_2.append(feature[feature['corr']==0][feature['power']<thresh])

# print("df_2 size: {} days".format(len(df_2)))
# df_3 = df_3.append(df)
# df_3 = df_3.append(df_2[:20])


# print("df_3 size: {} days".format(len(df_3)))
# print("Undefectable: {} days".format(len(feature)-len(df_3)))

# scores, cm, rf_model,X_test,y_test,test_date = model.random_forest_class(df_3)
# scores

# confusion_plot(cm[-1])
```
